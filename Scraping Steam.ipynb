```python
import requests
```

```python
import bs4
```

```python
headers = 'User-Agent: Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/11.1 Safari/605.1.15'
cookies = { 'birthtime': '283993201', 'mature_content': '1' }
```

```python
res = requests.get('https://store.steampowered.com/app/731490/Crash_Bandicoot_N_Sane_Trilogy/', headers, cookies=cookies)
```

```python
soup = bs4.BeautifulSoup(res.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
#Check if the content is age gated
for i in soup.find_all("h2"):
    if i.text.strip() == 'Please enter your birth date to continue:': #Age Gate check
        print(True)
    break
```

```python
#Print contents of the Contents div
content = soup.find("div", class_="page_content_ctn")
print(content.prettify())
```

```python
#Fetch the name form the content class
for content in soup.find_all("div", class_="page_content_ctn"):
    for i in content.select('span[itemprop="name"]'):
        print (i.text)
```

```python
#Fetch the current price from the content class
for content in soup.find_all("div", class_="page_content_ctn"):
    for items in content.select('meta[itemprop="price"]'):
        print(items)
```
